## Series
- A panda series is like column in a table
- It is a one dimensional array holding data of any type

```python
#creating series from list
a =[1,5,6]
myvar = pd.Series(a)
myvar
```

```python
#creating series from dictionary
data = {
    'a':10,
    'b':20,
    'c':30,
    'd':40,
    'e':50
}
series_from_dict = pd.Series(data)
series_from_dict
```

```python
#accesing elements of a series
myvar[0]
```

```python
# Performing operations on Series
series1 = pd.Series([1, 2, 3, 4, 5])
series2 = pd.Series([10, 20, 30, 40, 50])
series_sum = series1 + series2
print("Sum of two Series:")
print(series_sum)
```

<b> For a non-numerical Series object, describe() will give a simple summary of the number of unique values and most frequently occurring values</b>

```python
 s = pd.Series(['a', 'a', 'b', 'b', 'a', 'a', np.nan, 'c', 'd', 'a'])
s
```

<b>Note that on a mixed-type DataFrame object, describe() will restrict the summary to include only numerical columns or, if none are, only categorical columns:</b>

```python
frame = pd.DataFrame({'a': ['Yes', 'Yes', 'No', 'No'], 'b': range(4)})

frame
```

```python
frame.describe()
```

<b>This behaviour can be controlled by providing a list of types as include/exclude arguments. The special value all can also be used</b>
