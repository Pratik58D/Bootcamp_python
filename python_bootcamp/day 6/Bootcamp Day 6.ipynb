## Groupby in pandas
- It is used for grouping data according to the categories and applying a function to the categories
- Pandas dataframe.groupby() function is used to split the data into groups based on some criteria.


### Syntax: DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, squeeze=False, **kwargs)
<ul>
  <li><b>by:</b> mapping, function, str, or iterable</li>
  <li><b>axis:</b> int, default 0</li>
  <li><b>level:</b> If the axis is a MultiIndex (hierarchical), group by a particular level or levels</li>
  <li><b>as_index:</b> default True. Set to False if the result should NOT use the group labels as index</li>
  <li><b>sort:</b> Sort group keys. Get better performance by turning this off. Note this does not influence the order of observations within each group. groupby preserves the order of rows within each group.</li>
  <li><b>group_keys:</b> When calling apply, add group keys to index to identify pieces</li>
  <li><b>squeeze:</b> Reduce the dimensionality of the return type if possible, otherwise return a consistent type</li>
  <li><b>note:</b> by: is only required and others are optional</li>
</ul>

```python
#applying groupby() function to group the data on team value.

groupproduct = mydata.groupby('productid')  
"""
This line groups the entries in the DataFrame mydata
based on the unique values in the "productid" column.
This means that all rows with the same "productid" value 
are grouped together.
"""

'''
groupproduct.first() will return the first row of data
for each unique product ID in the DataFrame mydata.
'''
groupproduct.first()
```

```python
#it prints the all rows containing the given productid
groupproduct.get_group("0528881469")
```

```python
gkkk = mydata.groupby(["productid","userid"])
"""
gkkk.first() will return the first row of data for each unique 
combination of "userid" and "productid" in the DataFrame mydata. 
"""
gkkk.first()
```
